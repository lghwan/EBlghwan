# 데이터 가공하기

```python
# 데이터불러오기
import pandas as pd
filename='/content/sample_data/music_.xlsx'
df=pd.read_excel(filename)
df.head()
```

```python
df.columns
```

## 컬럼(변수) 삭제하기

```python
df.drop(['sortation'], axis=1, inplace=True)
df.columns
```

## 컬럼 이름 변경하기

```python
# 모든 컬럼 이름 변경하기
df.columns=['제목','이름','인원수','발매일','기획사','장르','유형']
df.head()
```

## 컬럼 생성하기

```python
df.dtypes
```

```python
# 새로운 컬럼 생성하기 : 년도,월,일
df['년도']=df['발매일'].dt.year
df['월']=df['발매일'].dt.month
df['일']=df['발매일'].dt.day
df.head()
```

```python
df
```

## 데이터 그룹핑

```python
# 장르별 가수인원수 출력하기
newdata=df.groupby('장르').count()
pd.DataFrame(newdata['인원수'])
```

```python
# 장르, 유형별  가수 인원 수 합계
newdata2=df.groupby(['장르','유형']).count()
pd.DataFrame(newdata2['인원수'])
```

## 결측데이터 확인

```python
# 데이터 불러오기
import pandas as pd
filename='/content/sample_data/train.csv'
df=pd.read_csv(filename)
df.head()
```

```python
# 결측데이터('탑승항구') 확인하기
df[df['Embarked'].isna()]
```

```python
# 컬럼별 결측데이터 개수 확인
df.isnull().sum()
```

```python
# 연습 : 'Age' 컬럼 결측데이터 확인하기

df[df['Age'].isna()]
```

```python
# 연습 : 'Cabin' 컬럼 결측데이터 확인하기

df[df['Cabin'].isna()]
```

## 결측데이터 대체 : 최빈값으로 채우기

```python
# 최빈값 찾기
# 'Embarked' 컬럼에 가장 빈번한 값은 S이고, S를 인덱싱 0으로 지정.
df['Embarked'].mode()
```

```python
#데이터프레임의 Embarked 컬럼의 결측데이터를 인덱싱 0 (값S로 대체)

df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
# 연습 - 결측데이터가 대체된 Embarked 컬럼의 결측데이터 수 확인

df['Embarked'].isna().sum()
```

```python
# 연습 - 결측데이터가 대체된 Embarked 컬럼의 결측데이터 수 확인

df.isnull().sum()
```
